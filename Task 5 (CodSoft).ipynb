```python
import matplotlib.pyplot as plt
import numpy as np

def plot_confusion_matrix(conf_matrix, labels):
    plt.figure(figsize=(8, 6))
    plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.get_cmap('Blues'))
    plt.title('Confusion Matrix')
    plt.colorbar()

    tick_marks = np.arange(len(labels))
    plt.xticks(tick_marks, labels, rotation=45)
    plt.yticks(tick_marks, labels)

    thresh = np.max(conf_matrix) / 2.
    for i in range(conf_matrix.shape[0]):
        for j in range(conf_matrix.shape[1]):
            plt.text(j, i, format(conf_matrix[i, j], 'd'),
                     horizontalalignment="center",
                     color="white" if conf_matrix[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Assuming you already have conf_matrix and LABELS defined
conf_matrix = np.array([[100, 10], [5, 50]])  # Example confusion matrix
LABELS = ['Normal', 'Fraud']  # Example labels
plot_confusion_matrix(conf_matrix, LABELS)
plt.show()
```

# Handling imbalanced- undersampling

```python
valid=df[df['Class']==0]
fraud=df[df['Class']==1]
valid.shape
```

```python
fraud.shape
```

```python
valid_sample=valid.sample(n=473)
valid_sample.shape
```

```python
new_data=pd.concat([valid_sample,fraud],ignore_index=True)
new_data['Class'].value_counts()
```

```python
new_data.head()
```

```python
x=new_data.drop('Class',axis=1)
y=new_data['Class']

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression

log=LogisticRegression()
log.fit(x_train,y_train)
```

```python
y_pred2=log.predict(x_test)

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

accuracy_score(y_test,y_pred2)
```

```python
recall_score(y_test,y_pred2)
```

```python
f1_score(y_test,y_pred2)
```

# Plot the POC Curve

```python
from sklearn import metrics

#define metrics
y_pred_proba=log.predict_proba(x_test)[::,1]
fpr,tpr,_=metrics.roc_curve(y_test,y_pred_proba)

#create ROC curve
plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# Calculate the AUC

```python
#define metrics
y_pred_proba=log.predict_proba(x_test)[::,1]
fpr,tpr,_=metrics.roc_curve(y_test,y_pred_proba)
auc=metrics.roc_auc_score(y_test,y_pred_proba)

#Create ROC curve
plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Postive Rate')
plt.legend(loc=4)
plt.show()
```
